TASK-5C: Please run the GaussianNB algorithm for 20 different values with respect to varr_smoothing parameter to find the best model. Then please visualize your findings via a lineplot (matplotlib) to illustrate the more appropriate values/ranges.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
digits.images.shape
```

```python
from sklearn.naive_bayes import GaussianNB                           
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from numpy import random

iris = sns.load_dataset('iris')
X_iris= iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=0)
```

```python
random.random
for i in range(20):
    print("{:.6f}".format(random.random()))
```

```python
Values=[0.534298,
0.992939,
0.774578,
0.318496,
0.516173,
0.468053,
0.798984,
0.443286,
0.783667,
0.913286,
0.319055,
0.408036,
0.977370,
0.459212,
0.994571,
0.395553,
0.995741,
0.726174,
0.814887,
0.353776]
```

```python
for i in range (20):
  V = Values[i]
  model = GaussianNB(var_smoothing=V)                       
  model.fit(Xtrain, ytrain)                  
  y_model = model.predict(Xtest)
  acc_scr = (accuracy_score(ytest, y_model)) 
  print(acc_scr)
  

plt.plot(Values)
plt.show()
```

```python
print("The best accuracy score: ", max(Values))
```
